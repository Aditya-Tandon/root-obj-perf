```python
pt_ranges = [(25, 100), (100, 200), (200, 300), (300, 400), (400, 500), (25, np.inf)]
eta_ranges = [(0, 0.5), (0.5, 1.5), (1.5, 2.0), (2.0, 2.4), (0, 2.4)]

collection_label = "UParT"
reco_jets = offline_upart_roc_events
tagger_name = upart_CONFIG["offline"]["tagger_name"]
var = "pt"
var_ranges = pt_ranges


def calculate_btag_roc_in_bins(gen_b_quarks, reco_jets, tagger_name, var, var_ranges):
    """
    Calculate b-tagging ROC curves in bins of a given variable.
    Parameters
    ----------
    gen_b_quarks : awkward.Array
        Generated b-quarks.
    reco_jets : awkward.Array
        Reconstructed jets.
    tagger_name : str
        Name of the b-tagging discriminator.
    var : str
        Variable to bin in (e.g., "pt" or "eta").
    var_ranges : list of tuple
        List of (min, max) tuples defining the bins for the variable.
    Returns
    -------
    dict
        Dictionary mapping bin labels to ROC curve points.
    """
    rocs = {}
    for var_min, var_max in var_ranges:
        var_mask = (abs(getattr(reco_jets, var)) >= var_min) & (
            abs(getattr(reco_jets, var)) < var_max
        )
        reco_jets_var_bin = reco_jets[var_mask]
        if len(reco_jets_var_bin) == 0:
            print(
                f"No jets found in {var} bin [{var_min}, {var_max}), skipping ROC calculation."
            )
            continue
        roc_points = calculate_roc_points(
            reco_jets_var_bin,
            get_purity_mask_cross_matched(gen_b_quarks, reco_jets_var_bin),
            tagger_name,
        )
        rocs[f"{var}_{var_min}_{var_max}"] = roc_points

    return rocs


rocs = calculate_btag_roc_in_bins(gen_b_quarks, reco_jets, tagger_name, var, var_ranges)
plot_roc_comparison(
    [
        (f"|{var}| [{var_min}, {var_max})", rocs[f"{var}_{var_min}_{var_max}"])
        for var_min, var_max in var_ranges
    ],
    working_point=0.1,
    title=rf"{collection_label} B-Tag ROC in |${var}$| Bins",
)
```

```python
import numpy as np
import awkward as ak
import matplotlib.pyplot as plt

from evaluation.roc import get_mistag_at_fixed_efficiency
from evaluation.jet_matching import calculate_roc_points, get_purity_mask_cross_matched
from plotting.roc_plots import plot_2d_performance_heatmap

# --- Configuration ---
var1, var1_ranges = "pt", pt_ranges
var2, var2_ranges = "eta", eta_ranges
target_eff = 0.10

# rocs_2d is keyed by (i, j) so the next grid-plot cell works directly.
rocs_2d = {}
auc_matrix = np.full((len(var1_ranges), len(var2_ranges)), np.nan, dtype=float)
mistag_matrix = np.full((len(var1_ranges), len(var2_ranges)), np.nan, dtype=float)
count_matrix = np.zeros((len(var1_ranges), len(var2_ranges)), dtype=float)

for i, (v1_min, v1_max) in enumerate(var1_ranges):
    for j, (v2_min, v2_max) in enumerate(var2_ranges):
        mask1 = (abs(getattr(reco_jets, var1)) >= v1_min) & (abs(getattr(reco_jets, var1)) < v1_max)
        mask2 = (abs(getattr(reco_jets, var2)) >= v2_min) & (abs(getattr(reco_jets, var2)) < v2_max)
        combined_mask = mask1 & mask2

        reco_jets_bin = reco_jets[combined_mask]
        n_bin = len(reco_jets_bin)
        count_matrix[i, j] = n_bin

        if n_bin == 0:
            continue

        purity_mask_bin = get_purity_mask_cross_matched(gen_b_quarks, reco_jets_bin)
        roc_points = calculate_roc_points(reco_jets_bin, purity_mask_bin, tagger_name)

        rocs_2d[(i, j)] = roc_points

        fpr, tpr, auc_val, _ = roc_points
        auc_matrix[i, j] = auc_val
        mistag_matrix[i, j] = get_mistag_at_fixed_efficiency(fpr, tpr, target_eff=target_eff)

# --- AUC heatmap ---
fig_auc = plot_2d_performance_heatmap(
    auc_matrix=auc_matrix,
    var1_ranges=var1_ranges,
    var2_ranges=var2_ranges,
    var1_name=var1,
    var2_name=var2,
    count_matrix=count_matrix,
    title=f"{collection_label} Tagger AUC in {var1} vs {var2}",
    vmin=0.5,
    vmax=1.0,
    cmap="viridis",
)
plt.show()

# --- Mistag@fixed-eff heatmap ---
mistag_vals = mistag_matrix[np.isfinite(mistag_matrix)]
mistag_vmax = float(np.nanpercentile(mistag_vals, 95)) if mistag_vals.size > 0 else 1.0

fig_mistag = plot_2d_performance_heatmap(
    auc_matrix=mistag_matrix,
    var1_ranges=var1_ranges,
    var2_ranges=var2_ranges,
    var1_name=var1,
    var2_name=var2,
    count_matrix=count_matrix,
    title=f"{collection_label} Mistag Rate at {target_eff*100:.0f}% efficiency in {var1} vs {var2}",
    vmin=0.0,
    vmax=mistag_vmax,
    cmap="viridis",
)
plt.show()
```

```python
# response and Scale Plots
# TODO: look at Prijith's work on jet energy scale and response to understand how muon/photon candles are used for this
def calculate_jet_responses(gen_particles, reco_objects, CONFIG=None):
    """
    Calculates pT and Energy response using vectorized
    Cross-Mutual Nearest Neighbor matching (1-to-1).

    Returns:
        gen_pt_flat:  The pT of the matched Gen particles (for the x-axis)
        pt_res_flat:  Reco / Gen pT response
        e_res_flat:   Reco / Gen Energy response
    """
    if CONFIG is None:
        with open("hh-bbbb-obj-config.json", "r") as config_file:
            CONFIG = json.load(config_file)

    gen_vec = gen_particles.vector[:, :, None]
    reco_vec = reco_objects.vector[:, None, :]

    # 2. Calculate DeltaR Matrix (All-to-All)
    delta_r_matrix = gen_vec.deltaR(reco_vec)

    # 3. Find Cross-Mutual Nearest Neighbors (Vectorized 1-to-1)

    # For every Gen particle, find index of closest Reco
    # shape: (events, n_gen)
    idx_closest_reco_to_gen = ak.argmin(delta_r_matrix, axis=2)

    # For every Reco jet, find index of closest Gen
    # shape: (events, n_reco)
    idx_closest_gen_to_reco = ak.argmin(delta_r_matrix, axis=1)

    # 4. Check the Cross-Match Condition
    # Get the index of the Gen particle that the closest Reco jet points to
    # We use the advanced indexing: array[indices]
    back_check_idx = idx_closest_gen_to_reco[idx_closest_reco_to_gen]

    # Create indices for comparison (0, 1, 2...)
    gen_indices = ak.local_index(gen_particles, axis=1)

    # A match is valid if:
    # 1. The indices match (Mutual agreement)
    # 2. The distance is within the cone size

    # Get the actual dR values for the closest matches
    min_dr_values = ak.min(delta_r_matrix, axis=2)

    is_mutual_match = (back_check_idx == gen_indices) & (
        min_dr_values < CONFIG["matching_cone_size"]
    )

    # 5. Extract the Matched Objects
    matched_gen = gen_particles[is_mutual_match]

    # We need to pull the specific Reco jets that matched these Gen particles
    # idx_closest_reco_to_gen contains the indices of the Reco jets we want
    # We apply the boolean mask to the INDICES first
    matched_reco_indices = idx_closest_reco_to_gen[is_mutual_match]

    # Now select those jets from the reco collection
    matched_reco = reco_objects[matched_reco_indices]

    # 6. Calculate responses
    # Formula: Reco / Gen

    # Pt response
    pt_res = matched_reco.vector.pt / matched_gen.vector.pt

    # Energy response
    matched_reco.vector["energy"] = np.sqrt(
        (matched_reco.vector.pt * np.cosh(matched_reco.vector.eta)) ** 2
        + (matched_reco.vector.mass) ** 2
    )
    e_res = matched_reco.vector.energy / matched_gen.vector.energy

    # Flatten arrays for easy plotting
    # We return the Gen Pt as the x-axis variable
    return (
        ak.to_numpy(ak.flatten(matched_gen.vector.eta)),
        ak.to_numpy(ak.flatten(matched_gen.vector.pt)),
        matched_reco,
        ak.to_numpy(ak.flatten(pt_res)),
        ak.to_numpy(ak.flatten(e_res)),
    )


def plot_response_vs_var(objects):
    """
    Bins the data by Gen pT and calculates the Mean and Width (StdDev)
    of the response in each bin.
    """
    plt.figure(figsize=(10, 8))
    for gen_var, response_values, bins, y_label, x_label, title in objects:
        bin_centers = 0.5 * (bins[1:] + bins[:-1])

        means = []
        widths = []
        errors = []  # Error on the width calculation

        # Digitize: Find which bin each event belongs to
        # indices 1 to len(bins)-1 are valid bins
        bin_indices = np.digitize(gen_var, bins)

        for i in range(1, len(bins)):
            # Select data for this specific bin
            data_in_bin = response_values[bin_indices == i]

            if len(data_in_bin) > 10:  # Require minimum stats
                # Calculate Mean (Bias)
                mu = np.mean(data_in_bin)

                # Calculate Width (response)
                # Standard Deviation is simple, but IQR/2 is more robust against tails
                sigma = np.std(data_in_bin)

                # Error on std dev estimate approx: sigma / sqrt(2N)
                err = sigma / np.sqrt(2 * len(data_in_bin))

                means.append(mu)
                widths.append(sigma)
                errors.append(err)
            else:
                means.append(np.nan)
                widths.append(np.nan)
                errors.append(0)

        # --- Plotting ---
        # Bottom Panel: response (Width)
        plt.subplot(2, 1, 2)
        plt.errorbar(
            bin_centers,
            widths,
            yerr=errors,
            fmt="o-",
            capsize=5,
            label=f"{y_label} response ($\sigma$)",
        )
        plt.ylabel(f"response")
        plt.title(title)
        plt.grid(True, linestyle="--", alpha=0.6)
        plt.legend()

        # Top Panel: Scale/Bias (Mean)
        plt.subplot(2, 1, 1)
        plt.errorbar(
            bin_centers, means, fmt="s--", capsize=5, label=f"{y_label} Scale (Mean)"
        )
        plt.axhline(1.0, color="black", linestyle="-", linewidth=1)
        plt.xlabel(x_label)
        plt.ylabel("Scale (Bias)")
        plt.grid(True, linestyle="--", alpha=0.6)
        plt.legend()

    plt.tight_layout()
    plt.show()


flat_gen_eta, flat_gen_pt, matched_reco, flat_pt_res, flat_e_res = (
    calculate_jet_responses(gen_b_quarks, reco_jets_offline)
)
(
    flat_gen_eta_l1ng,
    flat_gen_pt_l1ng,
    matched_reco_l1ng,
    flat_pt_res_l1ng,
    flat_e_res_l1ng,
) = calculate_jet_responses(gen_b_quarks, reco_jets_l1ng)
(
    flat_gen_eta_l1ext,
    flat_gen_pt_l1ext,
    matched_reco_l1ext,
    flat_pt_res_l1ext,
    flat_e_res_l1ext,
) = calculate_jet_responses(gen_b_quarks, reco_jets_l1ext)

pt_res_objects = [
    (
        flat_gen_pt,
        flat_pt_res,
        np.linspace(0, 500, 51),
        "Offline PNet Jet $p_T$",
        "Generated $p_T$ [GeV]",
        "Jet $p_T$ Scale and Response vs. Generated $p_T$",
    ),
    (
        flat_gen_pt_l1ng,
        flat_pt_res_l1ng,
        np.linspace(0, 500, 51),
        "L1NG Jet $p_T$",
        "Generated $p_T$ [GeV]",
        "Jet $p_T$ Scale and Response vs. Generated $p_T$",
    ),
    (
        flat_gen_pt_l1ext,
        flat_pt_res_l1ext,
        np.linspace(0, 500, 51),
        "L1Ext Jet $p_T$",
        "Generated $p_T$ [GeV]",
        "Jet $p_T$ Scale and Response vs. Generated $p_T$",
    ),
]
plot_response_vs_var(pt_res_objects)

eta_res_objects = [
    (
        flat_gen_eta,
        flat_e_res,
        np.linspace(-3, 3, 51),
        "Offline PNet Jet $p_T$",
        "Generated $\\eta$",
        "Jet $p_T$ Scale and Response vs. Generated $\\eta$",
    ),
    (
        flat_gen_eta_l1ng,
        flat_e_res_l1ng,
        np.linspace(-3, 3, 51),
        "L1NG Jet $p_T$",
        "Generated $\\eta$",
        "Jet $p_T$ Scale and Response vs. Generated $\\eta$",
    ),
    (
        flat_gen_eta_l1ext,
        flat_e_res_l1ext,
        np.linspace(-3, 3, 51),
        "L1Ext Jet $p_T$",
        "Generated $\\eta$",
        "Jet $p_T$ Scale and Response vs. Generated $\\eta$",
    ),
]
plot_response_vs_var(eta_res_objects)

pt_bins = np.linspace(0, 500, 51)
eta_bins = np.linspace(-3, 3, 51)
_, sigmas_pt_pnet, err_sigmas_pt_pnet, _, _ = plot_avg_attr_vs_var(
    reco_jets_offline,
    is_reco_jet_pure_offline,
    "PNetRegPtRawRes",
    "pt",
    pt_bins,
    "Generated Jet $p_T$ [GeV]",
    "PNetRegPtRawRes",
    "Offline Jet PNetRegPtRawRes vs. Generated $p_T$",
)

_, sigmas_eta_pnet, err_sigmas_eta_pnet, _, _ = plot_avg_attr_vs_var(
    reco_jets_offline,
    is_reco_jet_pure_offline,
    "PNetRegPtRawRes",
    "eta",
    np.linspace(-3, 3, 51),
    "Jet $\\eta$",
    "PNetRegPtRawRes",
    "Offline Jet PNetRegPtRawRes vs. $\\eta$",
)
_, sigmas_pt_upart, err_sigmas_pt_upart, _, _ = plot_avg_attr_vs_var(
    reco_jets_offline_upart,
    is_reco_jet_pure_offline_upart,
    "UParTAK4RegPtRawRes",
    "pt",
    pt_bins,
    "Generated Jet $p_T$ [GeV]",
    "UParTAK4RegPtRawRes",
    "Offline Jet UParTAK4RegPtRawRes vs. Generated $p_T$",
)
_, sigmas_eta_upart, err_sigmas_eta_upart, _, _ = plot_avg_attr_vs_var(
    reco_jets_offline_upart,
    is_reco_jet_pure_offline_upart,
    "UParTAK4RegPtRawRes",
    "eta",
    np.linspace(-3, 3, 51),
    "Jet $\\eta$",
    "UParTAK4RegPtRawRes",
    "Offline UParT Jet UParTAK4RegPtRawRes vs. $\\eta$",
)
```

```python
from evaluation.jet_matching import get_pure_jet_idxs_cross_matched


def get_eff_first_jet_pure(gen_b_quarks, reco_jets, tagger_name, n, k):
    """
    Calculates the efficiency of finding at least n pure jets in the top k jets
    ordered by pT and by the specified b-tagging score.

    gen_b_quarks: generated b-quarks
    reco_jets: reconstructed jets
    tagger_name: name of the b-tagging score to use for ordering
    n: Number of pure jets to find
    k: Max number of top jets to consider
    -----------
    Returns:
    eff_highest_pt_pure: Efficiency of highest pT jet being pure
    eff_highest_tag_pure: Efficiency of highest b-tagging score jet being pure
    more_than_n_eff_pt: List of efficiencies of finding at least n pure jets in top k jets ordered by pT
    more_than_n_eff_tag: List of efficiencies of finding at least n pure jets in top k jets ordered by b-tagging score
    -----------
    """
    pt_ordered = reco_jets[ak.argsort(reco_jets.vector.pt, ascending=False)]
    tag_ordered = reco_jets[ak.argsort(reco_jets[tagger_name], ascending=False)]

    pt_ordered_purity_idxs = get_pure_jet_idxs_cross_matched(gen_b_quarks, pt_ordered)
    tag_ordered_purity_idxs = get_pure_jet_idxs_cross_matched(gen_b_quarks, tag_ordered)

    # Eff highest pt jet is pure
    num_highest_pt_pure = ak.any(pt_ordered_purity_idxs == 0, axis=1)
    eff_highest_pt_pure = ak.sum(num_highest_pt_pure) / len(gen_b_quarks)
    print(f"Eff highest pt jet is pure: {eff_highest_pt_pure:.4f}")

    # Eff highest btag jet is pure
    num_highest_tag_pure = ak.any(tag_ordered_purity_idxs == 0, axis=1)
    eff_highest_tag_pure = ak.sum(num_highest_tag_pure) / len(gen_b_quarks)
    print(f"Eff highest {tagger_name} jet is pure: {eff_highest_tag_pure:.4f}")

    # Eff finding n pure jets in top k jets considered
    n = 4
    range_k = range(k + 1)

    more_than_n_eff_pt = []
    more_than_n_eff_tag = []
    for k in range_k:
        num_k_highest_pt_pure = (ak.num(pt_ordered_purity_idxs) == n) & (
            ak.all(pt_ordered_purity_idxs < k, axis=1)
        )
        eff_k_highest_pt_pure = ak.sum(num_k_highest_pt_pure) / len(gen_b_quarks)
        more_than_n_eff_pt.append(eff_k_highest_pt_pure)

        num_k_highest_tag_pure = (ak.num(tag_ordered_purity_idxs) == n) & (
            ak.all(tag_ordered_purity_idxs < k, axis=1)
        )
        eff_k_highest_tag_pure = ak.sum(num_k_highest_tag_pure) / len(gen_b_quarks)
        more_than_n_eff_tag.append(eff_k_highest_tag_pure)

    return (
        eff_highest_pt_pure,
        eff_highest_tag_pure,
        more_than_n_eff_pt,
        more_than_n_eff_tag,
    )


n = 4  # Number of b-jets to find
k = 15  # Max number of top jets to consider
range_k = range(k + 1)

pnet_jets = reco_jets_offline
upart_jets = reco_jets_offline_upart
l1ng_jets = reco_jets_l1ng
l1ext_jets = reco_jets_l1ext

(
    eff_highest_pt_pure_offline,
    eff_highest_btag_pure_offline,
    more_than_n_eff_pt_offline,
    more_than_n_eff_btag_offline,
) = get_eff_first_jet_pure(
    gen_b_quarks, pnet_jets, CONFIG["offline"]["tagger_name"], n, k
)
(
    eff_highest_pt_pure_upart,
    eff_highest_btag_pure_upart,
    more_than_n_eff_pt_upart,
    more_than_n_eff_btag_upart,
) = get_eff_first_jet_pure(
    gen_b_quarks, upart_jets, upart_CONFIG["offline"]["tagger_name"], n, k
)
(
    eff_highest_pt_pure_l1ng,
    eff_highest_btag_pure_l1ng,
    more_than_n_eff_pt_l1ng,
    more_than_n_eff_btag_l1ng,
) = get_eff_first_jet_pure(gen_b_quarks, l1ng_jets, CONFIG["l1ng"]["tagger_name"], n, k)
(
    eff_highest_pt_pure_l1ext,
    eff_highest_btag_pure_l1ext,
    more_than_n_eff_pt_l1ext,
    more_than_n_eff_btag_l1ext,
) = get_eff_first_jet_pure(
    gen_b_quarks, l1ext_jets, CONFIG["l1ext"]["tagger_name"], n, k
)


print(f"\nPlotting Top-N Jet Efficiencies for N = {n}...")
plt.figure(figsize=(10, 6))
plt.step(range_k, more_than_n_eff_pt_offline, where="mid", label="Offline PNet pT")
plt.step(range_k, more_than_n_eff_pt_upart, where="mid", label="Offline UPART pT")
plt.step(range_k, more_than_n_eff_pt_l1ng, where="mid", label="L1NG pT")
plt.step(range_k, more_than_n_eff_pt_l1ext, where="mid", label="L1EXT pT")
plt.xlabel("Top-k by pT (Number of Top Jets Considered)")
plt.ylabel(f"Probability of Finding at least {n} b-jets")
plt.title(f"Probability of Finding at least {n} b-jets vs. Top k Jets Considered")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()


plt.figure(figsize=(10, 6))
plt.step(range_k, more_than_n_eff_btag_offline, where="mid", label="Offline PNet BTag")
plt.step(range_k, more_than_n_eff_btag_upart, where="mid", label="Offline UPART BTag")
plt.step(range_k, more_than_n_eff_btag_l1ng, where="mid", label="L1NG BTag")
plt.step(range_k, more_than_n_eff_btag_l1ext, where="mid", label="L1EXT BTag")
plt.xlabel("Top-k by BTag (Number of Top Jets Considered)")
plt.ylabel(f"Probability of Finding at least {n} b-jets")
plt.title(f"Probability of Finding at least {n} b-jets vs. Top k Jets Considered")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
# di Higgs event reconstruction efficiency
from evaluation.dihiggs import find_gen_b_pairs_with_indices, R_hh_func
from plotting.dihiggs_plots import plot_ellipse


reco_jets = reco_jets_offline
tagger_name = CONFIG["offline"]["tagger_name"]
collection_label = "offline"
attr_to_sort_jets = tagger_name
top_k = 4  # only consider leading k jets for di-Higgs reconstruction


# # Select events with at least 4 gen b-quarks
# events_with_4b_mask = ak.num(gen_b_quarks) >= 4
# gen_b_quarks = gen_b_quarks[events_with_4b_mask]
# reco_jets = reco_jets[events_with_4b_mask]
# total_events = len(events_with_4b_mask)
# purity_mask = get_purity_mask_cross_matched(gen_b_quarks, reco_jets)


# from notes
# for pairing efficiency calculation, we need to match jets to gen b-quarks
reco_jets = reco_jets[
    ak.argsort(reco_jets[attr_to_sort_jets], ascending=False)
]  # pt_ordered
reco_jets = reco_jets[:, :top_k]  # only keep leading k jets for di-Higgs reconstruction
dr_matrix_reco_major = reco_jets[:, :, None].vector.deltaR(
    gen_b_quarks[:, None, :].vector
)
# Find closest GEN quark for each RECO jet. Values are Gen Indices (or None if N_gen=0)
idx_closest_gen_for_reco = ak.argmin(dr_matrix_reco_major, axis=2)  # events, n_reco
min_dr_reco_major = ak.fill_none(
    ak.min(dr_matrix_reco_major, axis=2), np.inf
)  # events, n_reco

dr_matrix_gen_major = gen_b_quarks[:, :, None].vector.deltaR(
    reco_jets[:, None, :].vector
)
# Find closest RECO jet for each GEN quark. Values are Reco Indices
idx_closest_reco_for_gen = ak.argmin(dr_matrix_gen_major, axis=2)  # events, n_gen

# We are indexing (Events, N_gen) using indices of shape (Events, N_reco)
# If idx_closest_gen_for_reco is None (no Gen quarks), this returns None safely.
back_check_reco_idx = idx_closest_reco_for_gen[idx_closest_gen_for_reco]

reco_idxs = ak.local_index(reco_jets, axis=1)
# fill_none(-1) handles the cases where N_gen=0 (where back_check is None)
# so they don't crash the comparison against integer Reco IDs.
pure_cross_matched = (ak.fill_none(back_check_reco_idx, -1) == reco_idxs) & (
    min_dr_reco_major < CONFIG["matching_cone_size"]
)

gmpr = idx_closest_gen_for_reco[pure_cross_matched]  # gen matched to pure reco jets


# di_Higgs matching and reconstruction
matched_jets = reco_jets[pure_cross_matched]
matched_jets = ak.drop_none(matched_jets)
at_least_4_mask = (
    ak.num(matched_jets) >= 4
)  # event level mask -> need to apply to everything else
matched_jets = matched_jets[
    at_least_4_mask
]  # need at least 4 matched jets for a di-Higgs candidate
print(
    f"Number of events with at least 4 matched jets: {len(matched_jets)} = {len(matched_jets)/len(reco_jets)*100:.2f}% of total events"
)

gmpr = gmpr[at_least_4_mask]
clean_gen_b_quarks = gen_b_quarks[at_least_4_mask]
clean_gen_particles = events.GenPart[at_least_4_mask]


_, _, true_h1_idxs, true_h2_idxs = find_gen_b_pairs_with_indices(
    gmpr, clean_gen_b_quarks, clean_gen_particles
)
# Sort the pairs internally so (2,3) is the same as (3,2)
true_h1_sorted = ak.sort(true_h1_idxs, axis=1)
true_h2_sorted = ak.sort(true_h2_idxs, axis=1)


top_4_jets = matched_jets[:, :4]
j0 = top_4_jets[:, 0]
j1 = top_4_jets[:, 1]
j2 = top_4_jets[:, 2]
j3 = top_4_jets[:, 3]

j0["gen_parent"] = gmpr[:, 0]
j1["gen_parent"] = gmpr[:, 1]
j2["gen_parent"] = gmpr[:, 2]
j3["gen_parent"] = gmpr[:, 3]

p1_c1 = gmpr[:, [0, 1]]  # candidate 1 from permutation 1
p1_c2 = gmpr[:, [2, 3]]  # candidate 2 from permutation 1

p2_c1 = gmpr[:, [0, 2]]  # candidate 1 from permutation 2
p2_c2 = gmpr[:, [1, 3]]  # candidate 2 from permutation 2

p3_c1 = gmpr[:, [0, 3]]  # candidate 1 from permutation 3
p3_c2 = gmpr[:, [1, 2]]  # candidate 2 from permutation 3

p11 = (j0, j1)  # candidate 1 from permutation 1
p12 = (j2, j3)  # candidate 2 from permutation 1

p21 = (j0, j2)  # candidate 1 from permutation 2
p22 = (j1, j3)  # candidate 2 from permutation 2

p31 = (j0, j3)  # candidate 1 from permutation 3
p32 = (j1, j2)  # candidate 2 from permutation 3

h_11_vec = p11[0].vector + p11[1].vector
h_12_vec = p12[0].vector + p12[1].vector
h_21_vec = p21[0].vector + p21[1].vector
h_22_vec = p22[0].vector + p22[1].vector
h_31_vec = p31[0].vector + p31[1].vector
h_32_vec = p32[0].vector + p32[1].vector

all_h1_mass = ak.concatenate(
    [h_11_vec.mass[:, None], h_21_vec.mass[:, None], h_31_vec.mass[:, None]], axis=1
)  # n, 3
all_h2_mass = ak.concatenate(
    [h_12_vec.mass[:, None], h_22_vec.mass[:, None], h_32_vec.mass[:, None]], axis=1
)
d_hh_all_pairs = abs(all_h1_mass - (125.0 / 120.0) * all_h2_mass) / np.sqrt(
    1 + (125.0 / 120.0) ** 2
)  # n, 3
min_d_hh_pair = ak.argmin(d_hh_all_pairs, axis=1)  # n

cond_0 = min_d_hh_pair == 0
cond_1 = min_d_hh_pair == 1

raw_h1_vec = ak.where(cond_0, h_11_vec, ak.where(cond_1, h_21_vec, h_31_vec))
raw_h2_vec = ak.where(cond_0, h_12_vec, ak.where(cond_1, h_22_vec, h_32_vec))

# Select the Jets of the winners
# H1 and H2 Jets
raw_h1_j1 = j0  # j0 is always the first jet of H1 in the scheme
raw_h1_j2 = ak.where(cond_0, j1, ak.where(cond_1, j2, j3))
raw_h2_j1 = ak.where(cond_0, j2, j1)  # Logic: if perm0(2), else j1
raw_h2_j2 = ak.where(cond_0, j3, ak.where(cond_1, j3, j2))
# Select the pairs corresponding to the winners
# algo_pair_A is H1 and algo_pair_B is H2
algo_pair_A = ak.where(cond_0[:, None], p1_c1, ak.where(cond_1[:, None], p2_c1, p3_c1))
algo_pair_B = ak.where(cond_0[:, None], p1_c2, ak.where(cond_1[:, None], p2_c2, p3_c2))


# --- Step 3: Apply pt Ordering (Swap H1 and H2 if H2 > H1) ---
is_h1_leading = raw_h1_vec.pt >= raw_h2_vec.pt
# 2. Sort the Jets (Swap the pairs)
# If True: Keep H1 as H1. If False: Swap H1 with H2.
leading_h_j1 = ak.where(is_h1_leading, raw_h1_j1, raw_h2_j1)
leading_h_j2 = ak.where(is_h1_leading, raw_h1_j2, raw_h2_j2)

subleading_h_j1 = ak.where(is_h1_leading, raw_h2_j1, raw_h1_j1)
subleading_h_j2 = ak.where(is_h1_leading, raw_h2_j2, raw_h1_j2)

algo_pair_leading = ak.where(is_h1_leading[:, None], algo_pair_A, algo_pair_B)
algo_pair_subleading = ak.where(is_h1_leading[:, None], algo_pair_B, algo_pair_A)
# 3. Sort them internally (so [3, 2] becomes [2, 3] for comparison)
algo_A_sorted = ak.sort(algo_pair_leading, axis=1)
algo_B_sorted = ak.sort(algo_pair_subleading, axis=1)

leading_h_vector = leading_h_j1.vector + leading_h_j2.vector
subleading_h_vector = subleading_h_j1.vector + subleading_h_j2.vector

# Calculate mass resolution if not L1 collection
# NOTE: PNetRegPtRawRes is RELATIVE (σ_pT / pT), so multiply by pT for absolute shift.
# Propagate each jet's uncertainty independently, then combine in quadrature.
if not collection_label.lower().startswith("l1"):
    # H1 (leading) — jet 1 contribution
    up_h11 = leading_h_j1.vector
    up_h11["pt"] = leading_h_j1.vector.pt * (1 + leading_h_j1.PNetRegPtRawRes)
    dn_h11 = leading_h_j1.vector
    dn_h11["pt"] = leading_h_j1.vector.pt * (1 - leading_h_j1.PNetRegPtRawRes)
    dm_h11 = (
        (up_h11 + leading_h_j2.vector).mass - (dn_h11 + leading_h_j2.vector).mass
    ) / 2

    # H1 (leading) — jet 2 contribution
    up_h12 = leading_h_j2.vector
    up_h12["pt"] = leading_h_j2.vector.pt * (1 + leading_h_j2.PNetRegPtRawRes)
    dn_h12 = leading_h_j2.vector
    dn_h12["pt"] = leading_h_j2.vector.pt * (1 - leading_h_j2.PNetRegPtRawRes)
    dm_h12 = (
        (leading_h_j1.vector + up_h12).mass - (leading_h_j1.vector + dn_h12).mass
    ) / 2

    sigma_mh1 = np.sqrt(dm_h11**2 + dm_h12**2)

    # H2 (subleading) — jet 1 contribution
    up_h21 = subleading_h_j1.vector
    up_h21["pt"] = subleading_h_j1.vector.pt * (1 + subleading_h_j1.PNetRegPtRawRes)
    dn_h21 = subleading_h_j1.vector
    dn_h21["pt"] = subleading_h_j1.vector.pt * (1 - subleading_h_j1.PNetRegPtRawRes)
    dm_h21 = (
        (up_h21 + subleading_h_j2.vector).mass - (dn_h21 + subleading_h_j2.vector).mass
    ) / 2

    # H2 (subleading) — jet 2 contribution
    up_h22 = subleading_h_j2.vector
    up_h22["pt"] = subleading_h_j2.vector.pt * (1 + subleading_h_j2.PNetRegPtRawRes)
    dn_h22 = subleading_h_j2.vector
    dn_h22["pt"] = subleading_h_j2.vector.pt * (1 - subleading_h_j2.PNetRegPtRawRes)
    dm_h22 = (
        (subleading_h_j1.vector + up_h22).mass - (subleading_h_j1.vector + dn_h22).mass
    ) / 2

    sigma_mh2 = np.sqrt(dm_h21**2 + dm_h22**2)

    rel_mass_res_h1 = sigma_mh1 / leading_h_vector.mass
    rel_mass_res_h2 = sigma_mh2 / subleading_h_vector.mass


r_hh_selection_mask = R_hh_func(leading_h_vector.mass, subleading_h_vector.mass) < 55.0


# Check A: Algo A matches Truth H1  AND  Algo B matches Truth H2 (truth ones are sorted by pt already)
match_direct = ak.all(algo_A_sorted == true_h1_sorted, axis=1) & ak.all(
    algo_B_sorted == true_h2_sorted, axis=1
)
pair_eff = ak.mean(match_direct)
print(f"Pairing Efficiency: {pair_eff:.2%}")

# Check B: H1 and H2 are swapped in the algo (i.e. Algo A matches Truth H2 and Algo B matches Truth H1)
# This is the swapped case where we assume the algo got the pairs ordered wrong but reconstructed the correct Higgs candidates.
# This is useful to see how often the algo gets the pairs swapped.
match_swapped = ak.all(algo_A_sorted == true_h2_sorted, axis=1) & ak.all(
    algo_B_sorted == true_h1_sorted, axis=1
)
# Calculate Efficiency
eff_swapped = ak.mean(match_swapped)
print(f"Pairing Efficiency with pairs swapped: {eff_swapped:.2%}")
print(f"Total Pairing Efficiency (including swapped): {pair_eff + eff_swapped:.2%}")

# Plotting the results
plt.hist2d(
    ak.to_numpy(leading_h_vector[r_hh_selection_mask].mass),
    ak.to_numpy(subleading_h_vector[r_hh_selection_mask].mass),
    bins=[np.linspace(0, 300, 61), np.linspace(0, 300, 61)],
    cmap="viridis",
)


plt.vlines(
    125, ymin=0, ymax=300, color="red", linestyle="--", label="Higgs Mass = 125 GeV"
)
plt.hlines(120, xmin=0, xmax=300, color="red", linestyle="--")
ax = plt.gca()
plot_ellipse(
    ax,
    125,
    120,
    110,
    96,
    0,
    edgecolor="yellow",
    facecolor="none",
    linestyle="--",
    label="R_hh = 55 GeV",
)
plt.title(f"Reconstructed Di-Higgs Mass Distribution for {collection_label}")
plt.xlabel("Leading Higgs Mass [GeV]")
plt.ylabel("Subleading Higgs Mass [GeV]")
plt.colorbar(label="Events")
plt.legend()
plt.show()

plt.hist2d(
    ak.to_numpy(rel_mass_res_h1[r_hh_selection_mask] * 100),
    ak.to_numpy(rel_mass_res_h2[r_hh_selection_mask] * 100),
    bins=[np.linspace(0, 20, 51), np.linspace(0, 20, 51)],
    cmap="plasma",
)
plt.xlabel("Relative Mass Resolution H1 [%]")
plt.ylabel("Relative Mass Resolution H2 [%]")
plt.title(f"Relative Mass Resolution for {collection_label}")
plt.colorbar(label="Events")
plt.show()
```

```python
from evaluation.dihiggs import (
    pair_from_4jets,
    find_gen_b_pairs_with_indices,
    reconstruct_dihiggs,
    compute_significance_at_luminosity,
    R_hh_func,
)

# === Helper functions for di-Higgs reconstruction across taggers ===
from matplotlib.patches import Ellipse
from data_pipeline.root_loading import load_and_prepare_data, apply_custom_cuts
import numpy as np
import awkward as ak
import gc

from evaluation.luminosity import load_physics_config

_physics = load_physics_config()
LUMINOSITY_FB = _physics["luminosity_fb"]
SIGNAL_XSEC_PB = _physics["signal_xsec_pb"]
N_GEN_SIGNAL = _physics["n_gen_signal"]


def _process_qcd_bin_for_tagger(
    qcd_events, collection_name, sort_attr, cfg, key, bin_weight
):
    """Process a single QCD bin for one tagger. Returns (lead_mass, sub_mass, hh_mass, n_events) or None.
    Only extracts mass arrays (numpy) to minimise memory footprint."""
    qcd_jets = apply_custom_cuts(qcd_events[collection_name], cfg, key)
    qcd_btag_sorted = qcd_jets[ak.argsort(qcd_jets[sort_attr], ascending=False)]
    has_4_tagged = ak.num(qcd_btag_sorted) >= 4
    qcd_4jets = qcd_btag_sorted[has_4_tagged][:, :4]
    n_events = int(ak.sum(has_4_tagged))
    if n_events == 0:
        return None
    q_lead, q_sub, q_hh = pair_from_4jets(qcd_4jets)
    # Extract masses as numpy immediately, discard full 4-vectors
    return (
        ak.to_numpy(q_lead.mass),
        ak.to_numpy(q_sub.mass),
        ak.to_numpy(q_hh.mass),
        n_events,
    )


def print_reco_summary(res):
    """Print a compact summary of a reconstruction result."""
    frac_sig = res["n_signal"] / res["n_total"] * 100 if res["n_total"] > 0 else 0
    # total_pe = (
    #     res["pair_eff"] + res["swap_eff"] if not np.isnan(res["pair_eff"]) else np.nan
    # )
    print(
        f"  [{res['label']}] Signal: {res['n_signal']} ({frac_sig:.1f}%)"
        f" | QCD bkg (unweighted): {res['n_qcd']}"
        f" | sum(weights): {np.sum(res['qcd_weights']):.1e}"
        # f" | Pairing: {res['pair_eff']:.1%} + {res['swap_eff']:.1%} = {total_pe:.1%}"
    )


# === Run reconstruction (Signal + real QCD) for all taggers ===
tagger_configs = [
    (
        "Offline PNet",
        reco_jets_offline,
        CONFIG,
        "offline",
        CONFIG["offline"]["tagger_name"],
    ),
    (
        "Offline UParT",
        reco_jets_offline_upart,
        upart_CONFIG,
        "offline",
        upart_CONFIG["offline"]["tagger_name"],
    ),
    ("L1NG", reco_jets_l1ng, CONFIG, "l1ng", CONFIG["l1ng"]["tagger_name"]),
    ("L1Ext", reco_jets_l1ext, CONFIG, "l1ext", CONFIG["l1ext"]["tagger_name"]),
]

# 1. Run signal reconstruction for all taggers (no QCD data in memory)
results = {}
for label, jets, cfg, key, sort_attr in tagger_configs:
    print(f"\nProcessing signal for {label}...")
    res = reconstruct_dihiggs(
        jets, gen_b_quarks, events, cfg, key, attr_to_sort_jets=sort_attr
    )
    res["label"] = label
    # Initialize QCD mass accumulators (numpy arrays, not full 4-vectors)
    res["_qcd_lead_mass"] = []
    res["_qcd_sub_mass"] = []
    res["_qcd_hh_mass"] = []
    res["_qcd_weights"] = []
    res["n_qcd"] = 0
    results[label] = res

# 2. Load each QCD bin ONCE and process all taggers before freeing.
#    Process high-pT bins first (smaller files, more likely to pass 4-jet b-tag cut).
all_collections = list(
    {cfg[key]["collection_name"] for _, _, cfg, key, _ in tagger_configs}
)
qcd_config = CONFIG.get("QCD_background", {})

# Sort bins by cross-section weight (ascending) so high-pT bins come first
qcd_bins_sorted = sorted(qcd_config.items(), key=lambda x: x[1]["weight"])

for bin_name, bin_cfg in qcd_bins_sorted:
    qcd_file_pattern = bin_cfg["file_pattern"]
    max_events_bin = bin_cfg.get("max_events", CONFIG.get("max_events", 1000))
    bin_weight = bin_cfg["weight"]
    n_gen_bin = bin_cfg["n_gen"]

    try:
        qcd_events = load_and_prepare_data(
            qcd_file_pattern,
            bin_cfg.get("tree_name", "Events"),
            all_collections,
            max_events=max_events_bin,
            correct_pt=True,
            CONFIG=CONFIG,
        )
    except Exception as e:
        print(f"  Error loading {bin_name}: {e}")
        continue

    if len(qcd_events) == 0:
        del qcd_events
        gc.collect()
        continue

    # Process all taggers from this single load
    for label, _, cfg, key, sort_attr in tagger_configs:
        collection_name = cfg[key]["collection_name"]
        result = _process_qcd_bin_for_tagger(
            qcd_events, collection_name, sort_attr, cfg, key, bin_weight
        )
        if result is not None:
            lead_m, sub_m, hh_m, n_events = result
            res = results[label]
            res["_qcd_lead_mass"].append(lead_m)
            res["_qcd_sub_mass"].append(sub_m)
            res["_qcd_hh_mass"].append(hh_m)
            res["_qcd_weights"].append(np.full(n_events, bin_weight, dtype=np.float64))
            res["n_qcd"] += n_events

    # Free raw QCD events immediately after all taggers processed this bin
    del qcd_events
    gc.collect()


# 3. Concatenate accumulated QCD masses and wrap with .mass attribute for downstream compat
class _MassArray:
    """Lightweight wrapper so downstream code can call .mass on QCD results."""

    __slots__ = ("mass",)

    def __init__(self, mass_np):
        self.mass = mass_np

    def __len__(self):
        return len(self.mass)


for label, res in results.items():
    if res["n_qcd"] > 0:
        res["qcd_lead"] = _MassArray(np.concatenate(res["_qcd_lead_mass"]))
        res["qcd_sub"] = _MassArray(np.concatenate(res["_qcd_sub_mass"]))
        res["qcd_hh"] = _MassArray(np.concatenate(res["_qcd_hh_mass"]))
        res["qcd_weights"] = np.concatenate(res["_qcd_weights"])
    else:
        res["qcd_lead"] = _MassArray(np.array([], dtype=np.float64))
        res["qcd_sub"] = _MassArray(np.array([], dtype=np.float64))
        res["qcd_hh"] = _MassArray(np.array([], dtype=np.float64))
        res["qcd_weights"] = np.array([], dtype=np.float64)
    # Remove temp accumulators
    del (
        res["_qcd_lead_mass"],
        res["_qcd_sub_mass"],
        res["_qcd_hh_mass"],
        res["_qcd_weights"],
    )
    # Compute luminosity-scaled signal weights
    # Build sigma → n_gen mapping for luminosity scaling in downstream cells
    res["sigma_to_ngen"] = {v["weight"]: v["n_gen"] for v in qcd_config.values()}
    print_reco_summary(res)

print(f"\nReconstruction complete for {len(results)} taggers.")
print(
    "QCD Background now read from explicit QCD simulation files with corresponding cross-section weights."
)
```

```python
from evaluation.roc import roc_from_scores
from evaluation.efficiency import btag_efficiency_vs_var, mistag_rate_vs_var

# === B-tagging performance: efficiency, ROC, and 2D AUC using QCD background ===
# Signal = pure (gen-matched) jets from signal ntuples
# Background = QCD jets (kinematic cuts only) loaded from QCD simulation files
# This complements cell 1 which uses impure signal-ntuple jets as background.

import gc
import numpy as np
import awkward as ak
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import roc_curve, auc
from data_pipeline.root_loading import (
    load_and_prepare_data,
    apply_custom_cuts as apply_custom_cuts_dl,
)

# ---------- configuration ----------
pt_ranges_btag = [
    (25, 100),
    (100, 200),
    (200, 300),
    (300, 400),
    (400, 500),
    (25, np.inf),
]
eta_ranges_btag = [(0, 0.5), (0.5, 1.5), (1.5, 2.0), (2.0, 2.4), (0, 2.4)]

tagger_defs = [
    ("Offline PNet", "offline", CONFIG, CONFIG["offline"]["tagger_name"]),
    ("Offline UParT", "offline", upart_CONFIG, upart_CONFIG["offline"]["tagger_name"]),
    ("L1NG", "l1ng", CONFIG, CONFIG["l1ng"]["tagger_name"]),
    ("L1Ext", "l1ext", CONFIG, CONFIG["l1ext"]["tagger_name"]),
]

# Map labels to the signal jets that already exist (kinematic-only for ROC)
signal_jets_kinematic = {
    "Offline PNet": apply_custom_cuts(
        events[CONFIG["offline"]["collection_name"]],
        CONFIG,
        "offline",
        kinematic_only=True,
    ),
    "Offline UParT": apply_custom_cuts(
        upart_events[upart_CONFIG["offline"]["collection_name"]],
        upart_CONFIG,
        "offline",
        kinematic_only=True,
    ),
    "L1NG": apply_custom_cuts(
        events[CONFIG["l1ng"]["collection_name"]], CONFIG, "l1ng", kinematic_only=True
    ),
    "L1Ext": apply_custom_cuts(
        events[CONFIG["l1ext"]["collection_name"]], CONFIG, "l1ext", kinematic_only=True
    ),
}

# Purity masks for signal jets (True = matched to gen b-quark)
signal_purity_masks = {}
for label in signal_jets_kinematic:
    signal_purity_masks[label] = get_purity_mask_cross_matched(
        gen_b_quarks, signal_jets_kinematic[label]
    )

# ---------- Load QCD jets (kinematic cuts only, once per bin for all collections) ----------
all_collections = list({cfg[key]["collection_name"] for _, key, cfg, _ in tagger_defs})
qcd_config = CONFIG.get("QCD_background", {})
qcd_bins_sorted = sorted(qcd_config.items(), key=lambda x: x[1]["weight"])
qcd_weights = {bin_name: bin_cfg["weight"] for bin_name, bin_cfg in qcd_bins_sorted}

# Accumulate flattened QCD tagger scores per label per (pt, eta) bin
# Also accumulate inclusive scores for the global ROC
qcd_scores_inclusive = {label: [] for label, *_ in tagger_defs}
qcd_pt_inclusive = {label: [] for label, *_ in tagger_defs}
qcd_eta_inclusive = {label: [] for label, *_ in tagger_defs}
qcd_weights_inclusive = {label: [] for label, *_ in tagger_defs}

print("Loading QCD jets for b-tag performance plots...")
for bin_name, bin_cfg in qcd_bins_sorted:
    qcd_file_pattern = bin_cfg["file_pattern"]
    max_events_bin = bin_cfg.get("max_events", CONFIG.get("max_events", 1000))

    try:
        qcd_events = load_and_prepare_data(
            qcd_file_pattern,
            bin_cfg.get("tree_name", "Events"),
            all_collections,
            max_events=max_events_bin,
            correct_pt=True,
            CONFIG=CONFIG,
        )
    except Exception as e:
        print(f"  Error loading {bin_name}: {e}")
        continue

    if len(qcd_events) == 0:
        del qcd_events
        gc.collect()
        continue

    for label, key, cfg, tagger_name in tagger_defs:
        collection_name = cfg[key]["collection_name"]
        # Kinematic cuts only (no b-tag cut) — mirrors signal jet selection
        qcd_jets = apply_custom_cuts_dl(
            qcd_events[collection_name], cfg, key, kinematic_only=True
        )
        scores = ak.to_numpy(ak.flatten(getattr(qcd_jets, tagger_name)))
        pts = ak.to_numpy(ak.flatten(qcd_jets.pt))
        etas = ak.to_numpy(ak.flatten(qcd_jets.eta))
        qcd_scores_inclusive[label].append(scores)
        qcd_pt_inclusive[label].append(pts)
        qcd_eta_inclusive[label].append(etas)
        qcd_weights_inclusive[label].append(
            np.full(len(scores), bin_cfg["weight"], dtype=np.float64)
        )

    del qcd_events
    gc.collect()

# Concatenate all QCD bins
for label in qcd_scores_inclusive:
    if qcd_scores_inclusive[label]:
        qcd_scores_inclusive[label] = np.concatenate(qcd_scores_inclusive[label])
        qcd_pt_inclusive[label] = np.concatenate(qcd_pt_inclusive[label])
        qcd_eta_inclusive[label] = np.concatenate(qcd_eta_inclusive[label])
        qcd_weights_inclusive[label] = np.concatenate(qcd_weights_inclusive[label])
    else:
        qcd_scores_inclusive[label] = np.array([])
        qcd_pt_inclusive[label] = np.array([])
        qcd_eta_inclusive[label] = np.array([])
        qcd_weights_inclusive[label] = np.array([], dtype=np.float64)

print("QCD jets loaded and flattened.")
for label in qcd_scores_inclusive:
    print(f"  {label}: {len(qcd_scores_inclusive[label]):,} QCD jets")


# ---------- Helper: ROC from separate sig/bkg arrays ----------
print("\n=== Inclusive ROC: Signal (gen-matched) vs QCD background ===")
roc_results_qcd = []
for label, key, cfg, tagger_name in tagger_defs:
    sig_jets = signal_jets_kinematic[label]
    pure_mask = signal_purity_masks[label]
    sig_scores = ak.to_numpy(ak.flatten(getattr(sig_jets[pure_mask], tagger_name)))
    bkg_scores = qcd_scores_inclusive[label]
    bkg_wts = qcd_weights_inclusive[label]
    roc_data = roc_from_scores(sig_scores, bkg_scores, bkg_weights=bkg_wts)
    roc_results_qcd.append((label, roc_data))
    print(
        f"  {label}: AUC = {roc_data[2]:.4f}  |  n_sig = {len(sig_scores):,}  |  n_bkg = {len(bkg_scores):,}"
    )

plot_roc_comparison(
    roc_results_qcd,
    working_point=0.1,
    title="B-Tag ROC: Signal (gen-matched) vs QCD Background",
)


# ---------- 2. B-tagging efficiency vs pT and eta ----------
# Efficiency = fraction of signal pure jets passing a b-tag cut, in bins of pT/eta
# Here we show the efficiency at the configured working point (b_tag_cut from config)
print("\n=== B-tagging efficiency vs pT and eta (at configured WP) ===")


pt_bins_plot = np.linspace(25, 500, 40)
eta_bins_plot = np.linspace(-2.4, 2.4, 40)
colors_btag = {
    "Offline PNet": "blue",
    "Offline UParT": "orange",
    "L1NG": "green",
    "L1Ext": "red",
}

fig, ax = plt.subplots()
for label, key, cfg, tagger_name in tagger_defs:
    b_tag_cut = cfg[key]["b_tag_cut"]
    bc, eff, err = btag_efficiency_vs_var(
        signal_jets_kinematic[label],
        signal_purity_masks[label],
        tagger_name,
        b_tag_cut,
        "pt",
        pt_bins_plot,
    )
    ax.errorbar(
        bc,
        eff,
        yerr=err,
        fmt="o-",
        label=f"{label} (cut={b_tag_cut:.3f})",
        color=colors_btag[label],
        markersize=3,
    )
ax.set_xlabel(r"Jet $p_T$ [GeV]")
ax.set_ylabel("B-tag Efficiency (signal pure jets)")
ax.set_title("B-tagging Efficiency vs $p_T$ (at configured WP)")
ax.set_ylim(0, 1.05)
ax.legend(fontsize=10)
ax.grid(True, linestyle="--", alpha=0.6)

# Efficiency vs eta
fig, ax = plt.subplots()
for label, key, cfg, tagger_name in tagger_defs:
    b_tag_cut = cfg[key]["b_tag_cut"]
    bc, eff, err = btag_efficiency_vs_var(
        signal_jets_kinematic[label],
        signal_purity_masks[label],
        tagger_name,
        b_tag_cut,
        "eta",
        eta_bins_plot,
    )
    ax.errorbar(
        bc,
        eff,
        yerr=err,
        fmt="o-",
        label=f"{label}",
        color=colors_btag[label],
        markersize=3,
    )
ax.set_xlabel(r"Jet $\eta$")
ax.set_ylabel("B-tag Efficiency (signal pure jets)")
ax.set_title(r"B-tagging Efficiency vs $\eta$ (at configured WP)")
ax.set_ylim(0, 1.05)
ax.legend(fontsize=10)
ax.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()


# ---------- 3. QCD mistag rate vs pT and eta ----------
print("\n=== QCD mistag rate vs pT and eta (at configured WP) ===")


fig, axes = plt.subplots(1, 2, figsize=(20, 7))

ax = axes[0]
for label, key, cfg, tagger_name in tagger_defs:
    b_tag_cut = cfg[key]["b_tag_cut"]
    bc, rate, err = mistag_rate_vs_var(
        qcd_scores_inclusive[label],
        qcd_pt_inclusive[label],
        b_tag_cut,
        pt_bins_plot,
        bkg_weights=qcd_weights_inclusive[label],
    )
    ax.errorbar(
        bc,
        rate,
        yerr=err,
        fmt="o-",
        label=f"{label} (cut={b_tag_cut:.3f})",
        color=colors_btag[label],
        markersize=3,
    )
ax.set_xlabel(r"Jet $p_T$ [GeV]")
ax.set_ylabel("QCD Mistag Rate")
ax.set_title("QCD Mistag Rate vs $p_T$ (at configured WP)")
ax.set_yscale("log")
ax.legend(fontsize=10)
ax.grid(True, linestyle="--", alpha=0.6, which="both")

ax = axes[1]
for label, key, cfg, tagger_name in tagger_defs:
    b_tag_cut = cfg[key]["b_tag_cut"]
    bc, rate, err = mistag_rate_vs_var(
        qcd_scores_inclusive[label],
        qcd_eta_inclusive[label],
        b_tag_cut,
        eta_bins_plot,
        bkg_weights=qcd_weights_inclusive[label],
    )
    ax.errorbar(
        bc,
        rate,
        yerr=err,
        fmt="o-",
        label=f"{label}",
        color=colors_btag[label],
        markersize=3,
    )
ax.set_xlabel(r"Jet $\eta$")
ax.set_ylabel("QCD Mistag Rate")
ax.set_title(r"QCD Mistag Rate vs $\eta$ (at configured WP)")
ax.set_yscale("log")
ax.legend(fontsize=10)
ax.grid(True, linestyle="--", alpha=0.6, which="both")

plt.tight_layout()
plt.show()


# ---------- 4. ROC in bins of pT ----------
print("\n=== ROC curves in pT bins (signal vs QCD) ===")
for label, key, cfg, tagger_name in tagger_defs:
    sig_jets = signal_jets_kinematic[label]
    pure_mask = signal_purity_masks[label]
    pure_jets = sig_jets[pure_mask]
    sig_scores_flat = ak.to_numpy(ak.flatten(getattr(pure_jets, tagger_name)))
    sig_pt_flat = ak.to_numpy(ak.flatten(pure_jets.pt))

    bkg_scores = qcd_scores_inclusive[label]
    bkg_pt = qcd_pt_inclusive[label]
    bkg_wts = qcd_weights_inclusive[label]

    roc_list = []
    for pt_lo, pt_hi in pt_ranges_btag:
        sig_mask = (sig_pt_flat >= pt_lo) & (sig_pt_flat < pt_hi)
        bkg_mask = (bkg_pt >= pt_lo) & (bkg_pt < pt_hi)
        roc_data = roc_from_scores(
            sig_scores_flat[sig_mask],
            bkg_scores[bkg_mask],
            bkg_weights=bkg_wts[bkg_mask],
        )
        roc_list.append((f"$p_T$ [{pt_lo}, {pt_hi})", roc_data))

    plot_roc_comparison(
        roc_list,
        working_point=0.1,
        title=f"{label}: B-Tag ROC in $p_T$ Bins (Signal vs QCD)",
    )


# ---------- 5. ROC in bins of eta ----------
print("\n=== ROC curves in |eta| bins (signal vs QCD) ===")
for label, key, cfg, tagger_name in tagger_defs:
    sig_jets = signal_jets_kinematic[label]
    pure_mask = signal_purity_masks[label]
    pure_jets = sig_jets[pure_mask]
    sig_scores_flat = ak.to_numpy(ak.flatten(getattr(pure_jets, tagger_name)))
    sig_eta_flat = np.abs(ak.to_numpy(ak.flatten(pure_jets.eta)))

    bkg_scores = qcd_scores_inclusive[label]
    bkg_eta = np.abs(qcd_eta_inclusive[label])
    bkg_wts = qcd_weights_inclusive[label]

    roc_list = []
    for eta_lo, eta_hi in eta_ranges_btag:
        sig_mask = (sig_eta_flat >= eta_lo) & (sig_eta_flat < eta_hi)
        bkg_mask = (bkg_eta >= eta_lo) & (bkg_eta < eta_hi)
        roc_data = roc_from_scores(
            sig_scores_flat[sig_mask],
            bkg_scores[bkg_mask],
            bkg_weights=bkg_wts[bkg_mask],
        )
        roc_list.append((f"|$\\eta$| [{eta_lo}, {eta_hi})", roc_data))

    plot_roc_comparison(
        roc_list,
        working_point=0.1,
        title=f"{label}: B-Tag ROC in |$\\eta$| Bins (Signal vs QCD)",
    )


# ---------- 6. 2D AUC heatmaps (pT x eta) per tagger ----------
print("\n=== 2D AUC heatmaps: pT vs |eta| (signal vs QCD) ===")


for label, key, cfg, tagger_name in tagger_defs:
    sig_jets = signal_jets_kinematic[label]
    pure_mask = signal_purity_masks[label]
    pure_jets = sig_jets[pure_mask]
    sig_scores_flat = ak.to_numpy(ak.flatten(getattr(pure_jets, tagger_name)))
    sig_pt_flat = ak.to_numpy(ak.flatten(pure_jets.pt))
    sig_eta_flat = np.abs(ak.to_numpy(ak.flatten(pure_jets.eta)))

    bkg_scores = qcd_scores_inclusive[label]
    bkg_pt = qcd_pt_inclusive[label]
    bkg_eta = np.abs(qcd_eta_inclusive[label])
    bkg_wts = qcd_weights_inclusive[label]

    auc_matrix = np.full((len(pt_ranges_btag), len(eta_ranges_btag)), np.nan)

    for i, (pt_lo, pt_hi) in enumerate(pt_ranges_btag):
        for j, (eta_lo, eta_hi) in enumerate(eta_ranges_btag):
            sig_mask = (
                (sig_pt_flat >= pt_lo)
                & (sig_pt_flat < pt_hi)
                & (sig_eta_flat >= eta_lo)
                & (sig_eta_flat < eta_hi)
            )
            bkg_mask = (
                (bkg_pt >= pt_lo)
                & (bkg_pt < pt_hi)
                & (bkg_eta >= eta_lo)
                & (bkg_eta < eta_hi)
            )
            if np.sum(sig_mask) < 10 or np.sum(bkg_mask) < 10:
                continue
            fpr, tpr, auc_val, _ = roc_from_scores(
                sig_scores_flat[sig_mask],
                bkg_scores[bkg_mask],
                bkg_weights=bkg_wts[bkg_mask],
            )
            auc_matrix[i, j] = auc_val

    df = pd.DataFrame(
        auc_matrix,
        index=[f"[{lo}, {hi})" for lo, hi in pt_ranges_btag],
        columns=[f"[{lo}, {hi})" for lo, hi in eta_ranges_btag],
    )

    plt.figure(figsize=(12, 9))
    sns.heatmap(
        df,
        annot=True,
        fmt=".4f",
        cmap="viridis",
        cbar_kws={"label": "AUC"},
        vmin=0.5,
        vmax=1.0,
    )
    plt.title(f"{label}: B-Tag AUC — $p_T$ vs |$\\eta$| (Signal vs QCD)", fontsize=14)
    plt.xlabel("|$\\eta$| Bin")
    plt.ylabel("$p_T$ Bin [GeV]")
    plt.tight_layout()
    plt.show()


# ---------- 7. Working point thresholds from QCD-based ROC ----------
# For each tagger, find the b-tag threshold at Tight/Medium/Loose mistag rates
TARGET_MISTAG_RATES = {"Tight": 0.001, "Medium": 0.01, "Loose": 0.1}

print("\n" + "=" * 70)
print("B-tag thresholds from QCD-based ROC curves")
print("=" * 70)
for label, roc_data in roc_results_qcd:
    fpr, tpr, auc_val, thresholds = roc_data
    print(f"\n{label}  (AUC = {auc_val:.5f})")
    for wp_name, target_mistag in TARGET_MISTAG_RATES.items():
        # Find first point where FPR >= target (FPR is sorted ascending from roc_curve)
        candidates = np.where(fpr >= target_mistag)[0]
        if len(candidates) == 0:
            print(f"  {wp_name:6s} (mistag={target_mistag:.1%}): no threshold found")
            continue
        idx = candidates[0]
        print(
            f"  {wp_name:6s} (mistag={target_mistag:.1%}): "
            f"thresh={thresholds[min(idx, len(thresholds)-1)]:.5f}  |  "
            f"TPR={tpr[idx]*100:.2f}%  |  "
            f"1/FPR={1/fpr[idx]:.1f}"
        )

print("\nB-tagging performance plots complete.")
```

```python
def get_genpart_flag_mask(genparts, flag_name):
    """
    Returns a boolean Awkward Array mask for a specific generator status flag.
    """
    # Mapping of flag names to their bit positions based on the ntuple docs
    flag_map = {
        "isPrompt": 0,
        "isDecayedLeptonHadron": 1,
        "isTauDecayProduct": 2,
        "isPromptTauDecayProduct": 3,
        "isDirectTauDecayProduct": 4,
        "isDirectPromptTauDecayProduct": 5,
        "isDirectHadronDecayProduct": 6,
        "isHardProcess": 7,
        "fromHardProcess": 8,
        "isHardProcessTauDecayProduct": 9,
        "isDirectHardProcessTauDecayProduct": 10,
        "fromHardProcessBeforeFSR": 11,
        "isFirstCopy": 12,
        "isLastCopy": 13,
        "isLastCopyBeforeFSR": 14,
    }

    if flag_name not in flag_map:
        raise ValueError(f"Flag '{flag_name}' not recognized.")

    bit_position = flag_map[flag_name]
    bit_value = 1 << bit_position

    # Perform bitwise AND across the entire array of UShort_t flags
    # If the result is greater than 0, the bit is active (True)
    return (genparts.statusFlags & bit_value) > 0


def decode_all_genpart_flags(genparts):
    """
    Takes the GenPart collection and returns an Awkward Record array
    containing boolean masks for every status flag simultaneously.
    """
    # Complete mapping from the documentation
    flag_map = {
        "isPrompt": 0,
        "isDecayedLeptonHadron": 1,
        "isTauDecayProduct": 2,
        "isPromptTauDecayProduct": 3,
        "isDirectTauDecayProduct": 4,
        "isDirectPromptTauDecayProduct": 5,
        "isDirectHadronDecayProduct": 6,
        "isHardProcess": 7,
        "fromHardProcess": 8,
        "isHardProcessTauDecayProduct": 9,
        "isDirectHardProcessTauDecayProduct": 10,
        "fromHardProcessBeforeFSR": 11,
        "isFirstCopy": 12,
        "isLastCopy": 13,
        "isLastCopyBeforeFSR": 14,
    }

    # Dictionary comprehension to create a boolean mask for every flag
    # We use genparts.statusFlags to access the bitwise UShort_t values
    decoded_masks = {
        flag_name: (genparts.statusFlags & (1 << bit_pos)) > 0
        for flag_name, bit_pos in flag_map.items()
    }

    # Zip the dictionary of arrays into a single Awkward array of records
    return ak.zip(decoded_masks)
```

```python
import awkward as ak
import vector
import fastjet


def build_gen_jets(events, jet_radius=0.4, min_pt=15.0):
    """
    Clusters stable, visible generator-level particles into AK4 GenJets.
    """
    print("Selecting stable, visible particles for clustering...")

    # 1. Require status == 1 (Stable particles)
    is_stable = events.GenPart.status == 1

    # 2. Exclude neutrinos (PDG IDs 12, 14, 16 and their antiparticles)
    abs_pdg = abs(events.GenPart.pdgId)
    is_visible = (abs_pdg != 12) & (abs_pdg != 14) & (abs_pdg != 16)

    # Apply the mask
    cluster_mask = is_stable & is_visible
    cluster_parts = events.GenPart[cluster_mask]

    # 3. Format the kinematics for FastJet using the vector library
    # We use the pt, eta, phi, and mass properties from the ntuple
    particles_p4 = ak.zip(
        {
            "px": cluster_parts.vector.px,
            "py": cluster_parts.vector.py,
            "pz": cluster_parts.vector.pz,
            "E": cluster_parts.vector.e,
            "user_index": ak.local_index(
                cluster_parts, axis=1
            ),  # CRITICAL: Must be named 'user_index'
        },
        with_name="Momentum4D",
    )

    print("Clustering particles with FastJet (Anti-kT, R=0.4)...")

    # 4. Define the Anti-kT algorithm with R=0.4
    jetdef = fastjet.JetDefinition(fastjet.antikt_algorithm, jet_radius)

    # 5. Perform the clustering
    cluster_sequence = fastjet.ClusterSequence(particles_p4, jetdef)

    # 6. Extract the inclusive jets, applying a minimum pT cut
    gen_jets = cluster_sequence.inclusive_jets(min_pt=min_pt)
    gen_jets["vector"] = ak.zip(
        {
            "pt": gen_jets.pt,
            "eta": gen_jets.eta,
            "phi": gen_jets.phi,
            "mass": gen_jets.mass,
        },
        with_name="Momentum4D",
    )
    print(f"Average GenJets per event: {ak.mean(ak.num(gen_jets)):.2f}")

    return gen_jets


def match_bquarks_to_genjets(true_b_quarks, gen_jets, max_dr=0.4):
    """
    Matches the true b-quarks to the newly clustered GenJets using Delta R.
    """
    # Calculate Delta R between every b-quark and every GenJet
    dr = true_b_quarks[:, :, None].vector.deltaR(gen_jets[:, None, :])

    # Find the index of the GenJet with the minimum Delta R for each b-quark
    closest_jet_idx = ak.argmin(dr, axis=2)
    min_dr = ak.min(dr, axis=2)

    # Ensure the match is within our cone radius (R=0.4)
    valid_match_mask = min_dr < max_dr

    # Get the actual matched GenJets (dropping quarks that didn't match a jet)
    matched_gen_jets = gen_jets[closest_jet_idx[valid_match_mask]]

    return matched_gen_jets


# Run the function
ak4_gen_jets = build_gen_jets(events)
matched_gen_pure_mask = get_purity_mask_cross_matched(gen_b_quarks, ak4_gen_jets)
matched_gen_jets = ak4_gen_jets[matched_gen_pure_mask]
pt_cuts = matched_gen_jets.pt > CONFIG["gen"]["pt_cut"]
eta_cuts = abs(matched_gen_jets.eta) < CONFIG["gen"]["eta_cut"]
matched_gen_jets = matched_gen_jets[pt_cuts & eta_cuts]
# matched_gen_jets = match_bquarks_to_genjets(gen_b_quarks, ak4_gen_jets)
print(f"Matched GenJets: {len(ak.flatten(matched_gen_jets))}")
```
